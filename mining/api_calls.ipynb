```python
import requests
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

# Data structure

```python
OPERAND = 'GET '
API_URL = 'https://www.okx.com'
CANDLESTICKS = '/api/v5/market/candles' #?instId={instId}'
CANDLESTICKS_HISTORY = '/api/v5/market/history-candles' #?instId={instId}&before={ts}&'
CANDLESTICKS_COLUMNS = ['ts', 'open', 'high', 'low', 'close', 'vol', 'volCcy']
```

```python
%%time
request_params = dict(
    instId = 'TONCOIN-USDT',
    limit = 300
)
headers = {
    'Accept-Encoding': 'gzip, deflate, br',
    'Connection': 'keep-alive'
}
test = requests.get(API_URL + CANDLESTICKS, params=request_params, headers=headers)
```

```python
test.json()['data']
```

```python
df = pd.DataFrame(test.json()['data'], columns=CANDLESTICKS_COLUMNS)
df['ts'] = pd.to_datetime(df['ts'], unit='ms')
df[CANDLESTICKS_COLUMNS[1:]] = df[CANDLESTICKS_COLUMNS[1:]].astype(float)
df
```

```python
fig = go.Figure(data=[go.Candlestick(x=df['ts'],
                                     open=df['open'],
                                     high=df['high'],
                                     low=df['low'],
                                     close=df['close'])])
fig.update_layout(template='plotly_dark')
fig.show()
```
